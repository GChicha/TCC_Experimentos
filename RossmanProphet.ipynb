```python
import pandas as pd
from fbprophet import Prophet
```

```python
data = pd.read_csv('train.csv', parse_dates=["Date"])
```

```python
m = Prophet()
```

```python
teste = data.rename(columns={"Date": "ds", "Sales": "y"})[data.Store == 1]
```

```python
aaa = teste.resample(on="ds", rule='W', ).sum()
```

```python
aaa['ds'] = aaa.index
```

```python
x = m.fit(aaa)
```

```python
bbb  = x.predict()[['ds','yhat']]
```

```python
bbb.set_index(bbb.ds).yhat.plot()
aaa.y.plot()
```

```python
x.plot_yearly()
```

```python
m.plot(fcst=m.predict())
```
