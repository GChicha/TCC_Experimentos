```python
from keras.layers import Dense, Activation
from keras.models import Sequential
from keras.optimizers import SGD

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import numpy as np

from statsmodels.datasets.co2 import load_pandas
```

```python
co2 = load_pandas().data
co2.interpolate(inplace=True)
```

```python
co2.plot()
```

```python
model = Sequential([
    Dense(32 , input_shape=(52,)),
    Activation('sigmoid'),
    Dense(32),
    Activation('sigmoid'),
    Dense(1),
    Activation('sigmoid')
])
```

```python
scaler = MinMaxScaler()
scaler.fit(co2)
train, test = train_test_split([x[0] for x in scaler.transform(co2)], shuffle=False)
```

```python
train_inputs = [(train[inicio:fim], train[fim]) for inicio, fim in zip(range(len(train) - 52), range(52, len(train)))]
X, y = np.array([x[0] for x in train_inputs]), np.array([y[1] for y in train_inputs])

test_inputs = [(test[inicio:fim], test[fim]) for inicio, fim in zip(range(len(test) - 52), range(52, len(test)))]
X_test, y_test = np.array([x[0] for x in test_inputs]), np.array([y[1] for y in test_inputs])
```

```python
sgd_optimizer = SGD(lr=.05, momentum=.5)
model.compile(optimizer=sgd_optimizer, loss='mean_squared_error')
```

```python
model.fit(X, y, epochs=500, batch_size=36)
```

```python
model.evaluate(X_test, y_test)
```

```python
plt.plot(scaler.inverse_transform(model.predict(X_test)))
```
